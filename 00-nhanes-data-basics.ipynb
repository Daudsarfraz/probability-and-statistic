# Data Visualization 

```python
# prepare some data
y = [6, 12,7, 2, 9,4, 5]
```

```python
import matplotlib  # plotting library
import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
plt.plot(y)
plt.show()
```

```python
# create a figure 
plt.figure(figsize=(10, 5)) # size of the figure

# add a line renderer with legend and line thickness
plt.plot(y,"*")
plt.plot(y)
# also need to make the figure readable -- very important from stats perspective!
plt.title("First Example by Dawood") # title of the graph
plt.xlabel("x vals (s)") # x-axis lable
plt.ylabel("y vals (m)") # y-axis lable
plt.grid() # show grids which are boxes in backgroung

plt.show()
```

```python
data['BMXWT'].size # show values or rows of that column|
```

```python
## Plotting Weights 
# create a figure 
plt.figure(figsize=(30, 15)) 

# add a line renderer with legend and line thickness
x =  range(data['BMXWT'].size)
y =  data['BMXWT']

#plt.scatter(x, y, marker='x', s=5) # x mean small cross already known by 
plt.scatter(x, y, marker = "x", s = 1)
# also need to make the figure readable -- very important from stats perspective!
plt.title("Weights of Individuals")
plt.xlabel("ID of Individual")
plt.ylabel("Weight (kg)")

plt.show()
```
